## 2. Series y Dataframes

Pandas tiene dos estructuras de datos principales: Series y DataFrames.

- **Series**: es un array unidimensional. Es similar a una columna en una tabla de Excel.
- **DataFrame**: es una estructura de datos bidimensional con filas y columnas con etiquetas. Es similar a una tabla de Excel.

### 2.1 Series

Para crear una Serie, puedes pasar una lista de elementos y Pandas crear치 una Serie con 칤ndices num칠ricos.

Dataframe son como tablas y series son columnas (listado de valores)

````python
# en pandas siempre los n칰meros de las filas. pd.Series() hace que la lista se vuelva en una columna. 
serie = pd.Series([1, 2, 3, 4, 5])
serie
````

### 2.2 DataFrames

Para crear un DataFrame, puedes pasar un diccionario de listas, donde cada clave es el nombre de la columna y cada valor es una lista de elementos.

````python
# rara vez se crea un dataset.

data = {
    'name': ['Mar칤a', 'Daniel', 'Marisa', 'David', 'Sabrina'],
    'age': [16, 7, 29, None, 33], 
    'city': ['Madrid', 'Madrid', 'M치laga', 'M치laga', 'Valladolid'],
    'height': [None, 1.20, 1.60, 2.01, 1.83]
}

df = pd.DataFrame(data)
df
````

Aunque lo m치s normal es crear DataFrames a partir de archivos CSV, Excel, bases de datos, etc. En este caso, vamos a guardar el DataFrame en un archivo CSV y posteriormente lo leeremos.

````python
CSV_PATH = 'data.csv'

# esto es para importar un csv.
# por lo general los dataframe se asignan a la variable df
df.to_csv(CSV_PATH, index=False)
df = pd.read_csv(CSV_PATH)
df
````

## 5. Indexaci칩n

Recordemos que el indexing consiste en seleccionar un subconjunto de datos de un DataFrame. Recordemos antes c칩mo era el dataframe original y veamos que Pandas nos permite indexar un DataFrame de diferentes maneras.

````python
df
````


**Por nombre de columna**: `df['column_name']`

````python
# obtienes el listado de nombres de la columna.
df['name']
````

Para invocar 2 columnas, se hace de la siguiente manera

````python
# Aqui se debe colocar corchete esto hace que est치s llamando a cierto intervalo.
df[['name', 'age']]
````

**Por posici칩n (칤ndice) de la fila**: `df.iloc[0]`


````python
# esto te brinda toda la informaci칩n de la fila en menci칩n.
df.iloc[0]
````

**Por etiqueta de la fila**: `df.loc['row_label']`


````python
df.loc[0]  # en este caso, como los indices son los mismos que las posiciones, `loc` es lo mismo que `iloc`
````

**Por etiqueta de fila y columna**: `df.loc['row_label', 'column_name']`

````python
# te indica el valor de de la posici칩n de la columna.
#df.iloc[posici칩n, 'nombre_de_columna']
df.loc[0, 'name']
````

**Por condici칩n**: `df[df['column_name'] > 0]`


````python

condition = df['age'] > 18

# si no indicado el df con los corchetes solo me saldr치 valores booleanos pero de esta manera se filtra.
df[condition]  # as칤 se filtran los datos
````

**Por valores m칤nimos y m치ximos**: `df['column_name'].idxmin()` y `df['column_name'].idxmax()`

````python
max_age_row_idx = df["age"].idxmax()
max_age_row = df.loc[max_age_row_idx]
max_age_row
````

````python
# Otra manera para mostrar que fila tiene la edad m치xima, pero solo mustra n칰mero
df["age"].idxmax()
````

````python
# para conocer la edad m치xima con numpy se hace asi se hace de esta manera.

import numpy as np

df["age"].max()
````

## 6. Data Cleaning

El data cleaning es un paso fundamental en cualquier an치lisis de datos. Pandas nos permite realizar operaciones de limpieza de datos de manera muy eficiente. **Es muy com칰n encontrarnos con valores nulos** en nuestros datasets. Pandas nos ofrece m칠todos para tratar estos valores nulos, como `dropna()`, `fillna()` y `isnull()`. Recordemos primero, el dataset original.

````python
df
````

Por ejemplo, para ver si hay valores nulos en un DataFrame, podemos usar los metodos `isnull()` 칩 `isna()`.

````python
# aqui te muestra la tabla y marca como True los valores null, tambi칠n se puede hacer con isna()
df.isnull()
````

Para **eliminar** las filas con valores nulos, podemos usar el m칠todo `dropna()`.

````python
df.dropna()
````

Para **rellenar** los valores nulos con un valor espec칤fico, podemos usar el m칠todo `fillna()`.

````python
# rellenas los valores y dentro de los parentesis indicas que valor quieres para rellenarlo.
df.fillna(0)  # rellena los valores nulos con 0
````

Estos m칠todos son aplicables tambi칠n a las Series (columnas de un DataFrame).

````python
df['age'].fillna(10)
````
